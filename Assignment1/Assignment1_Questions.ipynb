## Question 2: Extending the Hodgkin-Huxley Model

The Hodgkin–Huxley model accounts for sodium, potassium, and leak currents, but neurons also exhibit calcium dynamics that play a crucial role in excitability, synaptic plasticity, and bursting behavior. Calcium influx through voltage-gated channels not only affects the membrane potential directly, but also activates calcium-dependent potassium channels. Opening Ca²⁺ channels depolarizes the membrane (like Na⁺), but the entering Ca²⁺ also triggers the activation of Ca²⁺-dependent K⁺ channels that hyperpolarize the cell and shape firing patterns. In this problem, you will extend the Hodgkin–Huxley model to include these calcium dynamics.

### Auxiliary currents (at time $n$)

$$
I_{\exp}^{(n)} = g_{L,s}\,\Delta_T\,\exp\!\left(\frac{V_s^{(n)} - V_T}{\Delta_T}\right) \quad \text{(nA)}.
$$

$$
I_{\mathrm{ax},s}^{(n)} = g_c\,(V_d^{(n)} - V_s^{(n)}), \qquad
I_{\mathrm{ax},d}^{(n)} = g_c\,(V_s^{(n)} - V_d^{(n)}).
$$

$$
r_\infty(V) = \frac{1}{1+\exp\!\left(\frac{V - V_h}{k_h}\right)}, \qquad
r^{(n{+}1)} = r^{(n)} + \Delta t\,\frac{r_\infty\!\left(V_d^{(n)}\right) - r^{(n)}}{\tau_r},
$$

$$
I_h^{(n)} = g_h\,r^{(n)}\,(V_d^{(n)} - E_h).
$$

$$
m_\infty(V) = \frac{1}{1+\exp\!\left(-\frac{V - V_p}{k_p}\right)}, \qquad
I_{CaP}^{(n)} = g_{CaP}\,m_\infty\!\left(V_d^{(n)}\right)\,\left(V_d^{(n)} - E_{Ca}\right).
$$
*(Disable by setting $g_{CaP}=0$.)*

### Euler updates (discrete with step $\Delta t$)

**1 Update $r$**
$$
r^{(n{+}1)} = r^{(n)} + \Delta t\,\frac{r_\infty\!\left(V_d^{(n)}\right) - r^{(n)}}{\tau_r}.
$$

**2 Refractory (soma)**
If $n_{\mathrm{ref}}^{(n)} > 0$:  
$$
V_s^{(n{+}1)} = V_r,\qquad n_{\mathrm{ref}}^{(n{+}1)} = n_{\mathrm{ref}}^{(n)} - 1.
$$
(Still update dendrite in step 3b.)

**3a Soma (if not refractory)**
$$
V_s^{(n{+}1)} = V_s^{(n)} + \frac{\Delta t}{C_s}\left[-g_{L,s}\left(V_s^{(n)}-E_L\right) + I_{\exp}^{(n)} - w^{(n)} + I_{\mathrm{ax},s}^{(n)} + I_s^{\mathrm{ext}}\right].
$$

**3b Dendrite (always)**
$$
V_d^{(n{+}1)} = V_d^{(n)} + \frac{\Delta t}{C_d}\left[-g_{L,d}\left(V_d^{(n)}-E_L\right) - I_{\mathrm{ax},d}^{(n)} - I_h^{(n)} + I_d^{\mathrm{ext}} - I_{CaP}^{(n)}\right].
$$

**4 Adaptation (soma)**
$$
w^{(n{+}1)} = w^{(n)} + \Delta t\,\frac{a\,(V_s^{(n)} - E_L) - w^{(n)}}{\tau_w}.
$$

**5 Spike test & reset (soma)**
